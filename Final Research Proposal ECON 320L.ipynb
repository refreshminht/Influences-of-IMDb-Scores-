# Testing for Assumptions (ex. OVB, Multicollinearity, Heteroskedasticity)

### Omitted Variable Bias 

We are constrained by the limits of our dataset. There are not many quantitative variables that we have in use for our analysis. We do have complete data on the quantitative independent variables in our possession, which are: duration, number of written reviews, and number of votes. 

Though there may be other quantitative variables that could affect a movie's IMDB rating (ex: # of years experience for lead actor), these are not contained within our dataset, and there is no way to account for it. Still, it is good to keep in mind the limitations of our dataset and discuss it when talking about our final results

### Multicollinearity 

If multicollinearity fails between variables, then one of the variables must be taken out. Multicollinearity is an assuption where one variable cannot be a linear transformation of another variable. The only two variables that would even need to be tested for this are between "number of written reviews" and "number of votes". Though I am sure that these variables are correlated (and even highly so), all we need to ensure is that one is not a linear transformation of the other. 

```python
# Cleaning data to get rid of rows that have no values for either # of reviews

movie1 = movie.dropna(axis=0, subset=['num_user_for_reviews'])

user_votes = movie1['num_voted_users']
user_reviews = movie1['num_user_for_reviews']

corr, _ = pearsonr(user_votes, user_reviews)
print('Correlation Between # of votes and written reviews: %.3f' % corr)
```

So, looking at these results, there is still a high correlation between the number of votes and number of written reviews. This was expected. However, there is not a direct linear relationship (which would imply a correlation of 1) between these two variables. Therefore, it is still valid to place within our dataset.

### Heteroskedasticity

Regressions are more powerful when they are homoskedastic. Homoskedasticity is an important assumption that we employ. In modern times, since the advent of Robust White standard errors, we can just make a regression table with White standard errors and the default standard error, just so we can see the difference. 

If we only used default standard errors, then they would be too small for a heteroskedastic model (which we have). This is why we use White standard errors -- they help correct for heteroskedasticity in models that are not homoskedastic. 

Below, when we do the regression, we will demonstrate both versions -- White Standard Error and Regular Standard Error.

# Methodology/Empirical Model

For clarification, we will be running two regressions. One with default standard errors, and one with Robust White standard errors. Both can provide us with valuable information about how our data changes when accounting for heteroskedasticity.

```python
# Regression with regular standard errors

mod = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews', 
              data=movie).fit()

#regression w/ Documentary
mod1 = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews + isDocumentary', 
              data=movie).fit()

#regression w/ English
mod2 = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews +isEnglish', 
              data=movie).fit()

#regression w/ USA
mod3 = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews +isUSA', 
              data=movie).fit()

#regression w/ English and USA
mod4 = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews +isEnglish + isUSA', 
              data=movie).fit()

#Regression w/ all
mod5 = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews + isDocumentary+isEnglish + isUSA', 
              data=movie).fit()

```

```python
# Regression with White Standard Errors 

mod_w = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews', 
              data=movie).fit(cov_type = 'HC0')
#regression w/ Documentary
mod1_w = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews + isDocumentary', 
              data=movie).fit(cov_type ='HC0')

#regression w/ English
mod2_w = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews +isEnglish', 
              data=movie).fit(cov_type ='HC0')

#regression w/ USA
mod3_w = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews +isUSA', 
              data=movie).fit(cov_type ='HC0')

#regression w/ English and USA
mod4_w = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews +isEnglish + isUSA', 
              data=movie).fit(cov_type ='HC0')

#Regression w/ all
mod5_w = smf.ols(formula='imdb_score ~ duration + num_voted_users + num_user_for_reviews + isDocumentary+isEnglish + isUSA', 
              data=movie).fit(cov_type ='HC0')
```

Our thought process that we took when choosing our model was how we could accurately show the data and the best way to find a statistically significant relationship. First, we analyzed the relationship between IMDb score and the numerical variables (duration, number user for reviews, voted users) through scatterplots because scatterplots help visualize the data as well as if theres any trend between the independent and dependent variables. We created dummy variables to cover any influence the none-numerical variables may have had. Since these variables arent numerical, we used box plots to find trends within IMDb scores compared to variables such as language and country. We believe that this approach to our models helps identify trends in the data in a clear way.

First, we used the duration variable to see if the length of the movie affects the IMDb score. We included this variable because its numerical and its possible that movie length has an effect on a movie - for example, if the movie is too long, the critic might get bored and give it a lower score, or if the movie is too short, the critic may feel unsatisfied and also give it a lower score. 
Next, we used the number of voted users vs. IMDb score to see if the number of voters affect the IMDb score at all. Less voters on a particular movie  would make each vote have greater weight in the final score, whereas the opposite would happen for a movie with lots of votes. We were interested to see how this numerical difference affects ratings. 
Third, we used the number of reviews vs IMDb score. This reasoning for using this variable is similar to the reasons we had for using the number of voted users vs. IMDb score, but there is a difference in voting and reviews in that voting just submits a rating out of 5 stars while reviews require a person to write their thoughts about the movie. We were interested to see if the number of reviews has any impact on the IMDb score because more reviews could mean that either the movie was very good or bad that it motivated more people to write about it. 
For our first dummy variables, we used language. Language has a possible effect on IMDb because higher ratings could be applied to Western and English movies. We believe this due to the fact that much of the film industry is dominated by white people and so far in history, critics have not been as receptive to movies that are not English (as seen as Parasite being the first non-English movie to win an Oscar, which was only recently). 
Our second dummy variable was the country the movie was made in. We applied similar logic from using the language variable to the country variable. We also wanted to account for possible outliers in our data based on countries that produce a majority or a minority of movies analyzed.


The director name variable could be important but we decided not to include it. This is because a celebrity director could have an impact on IMDb scores based on awareness alone, whereas a no-name director would garner less attraction around their movie and thus, less ratings to contribute to their IMDb score. Looking here, it seems that while the standard errors were definitely higher with White Standard Errors, it does not affect any of the actual coefficients, nor does it make the p-value high enough to actually be put in doubt.
